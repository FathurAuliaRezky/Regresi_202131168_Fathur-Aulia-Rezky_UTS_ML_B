## Import Library

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
```

Pada bagian ini, dilakukan import library yang dibutuhkan untuk melakukan preprocessing data, melatih data, dan menampilkan akurasi dan visualisasi model regresi. Library yang digunakan adalah pandas untuk membaca dataset, matplotlib untuk menampilkan visualisasi model, dan sklearn untuk melakukan preprocessing data dan melatih model regresi.

## Load dataset

```python
df = pd.read_csv('R02_rice_field.csv')
```

Pada bagian ini, dilakukan load dataset menggunakan pandas. Dataset yang digunakan adalah file R02_rice_field.csv yang telah disediakan.

## Pisahkan atribut dan label

```python
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

Pada bagian ini, dilakukan pemisahan dataset menjadi atribut (X) dan label (y). Atribut adalah kolom produksi padi, sedangkan label adalah kolom luas panen.

## Bagi dataset menjadi data latih dan data uji

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

Pada bagian ini, dilakukan pembagian dataset menjadi data latih dan data uji menggunakan train_test_split. Data latih digunakan untuk melatih model regresi, sedangkan data uji digunakan untuk menguji akurasi model.


## Lakukan scaling pada data

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

Pada bagian ini, dilakukan preprocessing data dengan melakukan scaling menggunakan StandardScaler. Scaling dilakukan untuk mengubah skala data agar memiliki rentang nilai yang sama.

## Latih model regresi

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

Pada bagian ini, dilakukan pelatihan model regresi menggunakan LinearRegression. Model regresi digunakan untuk memprediksi nilai luas panen berdasarkan nilai produksi padi.

## Hitung akurasi model

```python
accuracy = regressor.score(X_test, y_test)
print("Akurasi model: {:.2f}%".format(accuracy*100))
```

Pada bagian ini, dilakukan penghitungan akurasi model menggunakan metode score. Akurasi model menunjukkan seberapa baik model dapat memprediksi nilai luas panen berdasarkan nilai produksi padi.

## Visualisasikan model

```python
plt.scatter(X_train, y_train, color='red')
plt.plot(X_train, regressor.predict(X_train), color='blue')
plt.title('Produksi Padi vs Luas Panen (Training set)')
plt.xlabel('Produksi Padi')
plt.ylabel('Luas Panen')
plt.show()
```

Pada bagian ini, dilakukan visualisasi model menggunakan matplotlib. Visualisasi model menunjukkan hubungan antara nilai produksi padi dan luas panen. Garis biru menunjukkan hasil prediksi model regresi, sedangkan titik merah menunjukkan data latih.
